```python
import numpy as np  # linear algebra
import pandas as pd  # data processing, CSV file I/O (e.g., pd.read_csv)
import nltk

# List of items for frequency distribution
items = ['apple', 'apple', 'kiwi', 'cabbage', 'cabbage', 'potato']

# Frequency distribution of the items
fdist = nltk.FreqDist(items)
print(fdist)

# Conditional frequency distribution example
c_items = [('F', 'apple'), ('F', 'apple'), ('F', 'kiwi'), ('V', 'cabbage'), ('V', 'cabbage'), ('V', 'potato')]

# Create a conditional frequency distribution
cfd = nltk.ConditionalFreqDist(c_items)

# Display the conditions in the conditional frequency distribution
print("Conditions in the conditional frequency distribution:", cfd.conditions())

# Plot the conditional frequency distribution
cfd.plot()

# Accessing the 'V' condition
print("Conditional frequency distribution for 'V':")
print(cfd['V'])
```

```python
import nltk
from nltk.corpus import brown

# Create a Conditional Frequency Distribution (cfd) for the brown corpus
cfd = nltk.ConditionalFreqDist(
    [(genre, word) for genre in brown.categories() for word in brown.words(categories=genre)]
)

# Display the Conditional Frequency Distribution (cfd)
print(cfd)

# List all conditions (categories of the Brown corpus)
print("Conditions:", cfd.conditions())

# Tabulate the frequency of specific words (samples) under certain categories (conditions)
cfd.tabulate(conditions=['government', 'humor', 'reviews'], samples=['leadership', 'worship', 'hardship'])

# Plot the frequency of specific words (samples) under certain categories (conditions)
cfd.plot(conditions=['government', 'humor', 'reviews'], samples=['leadership', 'worship', 'hardship'])

# Tabulate the cumulative frequency of specific words (samples) under certain categories (conditions)
cfd.tabulate(conditions=['government', 'humor', 'reviews'], samples=['leadership', 'worship', 'hardship'], cumulative=True)

# Get the frequency distribution for the 'news' category
news_fd = cfd['news']

# Get the most common words in the 'news' category
print("Most common words in 'news' category:", news_fd.most_common(3))

# Get the frequency of the word 'the' in the 'news' category
print("Frequency of 'the' in 'news' category:", news_fd['the'])
```

```python
import nltk
from nltk.corpus import names

# Create a list of tuples where each tuple contains the gender ('male' or 'female') and the last letter of each name
nt = [(fid.split('.')[0], name[-1]) for fid in names.fileids() for name in names.words(fid)]

# Create a Conditional Frequency Distribution (cfd2) for the list of tuples (gender, last letter)
cfd2 = nltk.ConditionalFreqDist(nt)

# Get the frequency of the letter 'a' in female names
print(cfd2['female']['a'])

# Get the frequency of the letter 'a' in male names
print(cfd2['male']['a'])

# Compare the frequency distribution of female and male names
print(cfd2['female'] > cfd2['male'])

# Tabulate the frequencies of the letters 'a' and 'e' in male and female names
cfd2.tabulate(samples=['a', 'e'])

# Plot the frequency distribution of the last letters in male and female names
cfd2.plot()
```

```python
# Import necessary libraries
import math
import numpy as np
from numpy.random import randn
from statsmodels.stats.weightstats import ztest

# Generate a random array of 50 numbers having mean 110 and sd 15
# Similar to the IQ scores data we assume above
mean_iq = 110
sd_iq = 15 / math.sqrt(50)  # Standard deviation adjusted for sample size
alpha = 0.05  # Significance level for the test
null_mean = 100  # Null hypothesis mean

# Generate random data using the calculated standard deviation
data = sd_iq * randn(50) + mean_iq

# Print mean and standard deviation of the generated data
print('Mean=%.2f, Standard Deviation=%.2f' % (np.mean(data), np.std(data)))

# Perform the z-test. We test whether the sample mean is larger than the null hypothesis mean
ztest_Score, p_value = ztest(data, value=null_mean, alternative='larger')

# Print the z-score and p-value
print(f'Z-test Score: {ztest_Score:.4f}, P-value: {p_value:.4f}')

# Compare p-value with alpha to decide whether to reject the null hypothesis
if p_value < alpha:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
# Importing the required libraries and packages
import numpy as np
from scipy import stats

# Defining two random distributions
# Sample Size
N = 10

# Gaussian distributed data with mean = 2 and var = 1
x = np.random.randn(N) + 2

# Gaussian distributed data with mean = 0 and var = 1
y = np.random.randn(N)

# Calculating the variance to get the standard deviation
var_x = x.var(ddof=1)  # Sample variance for x
var_y = y.var(ddof=1)  # Sample variance for y

# Standard Deviation
SD = np.sqrt((var_x + var_y) / 2)

print("Standard Deviation =", SD)

# Calculating the T-Statistics
tval = (x.mean() - y.mean()) / (SD * np.sqrt(2 / N))

# Degrees of freedom
dof = 2 * N - 2

# p-value after comparison with the T-Statistics
pval = 1 - stats.t.cdf(tval, df=dof)

print("t = " + str(tval))
print("p = " + str(2 * pval))

# Cross Checking using the internal function from SciPy Package
tval2, pval2 = stats.ttest_ind(x, y)

print("t (SciPy) = " + str(tval2))
print("p (SciPy) = " + str(pval2))
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Sample Data: Create a sample DataFrame `df`
data = {
    'A': [23, 21, 22, 23, 24],
    'B': [31, 33, 32, 30, 29],
    'C': [43, 42, 44, 45, 43],
    'D': [15, 14, 13, 16, 17]
}

df = pd.DataFrame(data)

# Perform ANOVA using scipy.stats
fvalue, pvalue = stats.f_oneway(df['A'], df['B'], df['C'], df['D'])
print(f"F-value: {fvalue}, p-value: {pvalue}")

# Reshape the data for ANOVA using statsmodels
df_melt = df.melt(var_name='treatments', value_name='value')

# Fit the model using Ordinary Least Squares (OLS)
model = ols('value ~ C(treatments)', data=df_melt).fit()

# Perform ANOVA using statsmodels
anova_table = sm.stats.anova_lm(model, typ=2)
print("\nANOVA Table:\n", anova_table)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib as mpl

# Load the first dataset (anti-diabetic drug sales)
df_a10 = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/a10.csv', parse_dates=['date'])
print("First 5 rows of the anti-diabetic drug sales dataset:")
print(df_a10.head())

# Load the second dataset (market arrivals in Mumbai)
df_mumbai = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/MarketArrivals.csv')
df_mumbai = df_mumbai.loc[df_mumbai.market == 'MUMBAI', :]
print("\nFirst 5 rows of the market arrivals dataset for Mumbai:")
print(df_mumbai.head())

# Load the third dataset (anti-diabetic drug sales again, with date as index)
df_a10_indexed = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/a10.csv', parse_dates=['date'], index_col='date')

# Draw Plot Function
def plot_df(df, x, y, title="", xlabel='Date', ylabel='Value', dpi=100):
    plt.figure(figsize=(16, 5), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()

# Plot the anti-diabetic drug sales time series
plot_df(df_a10_indexed, x=df_a10_indexed.index, y=df_a10_indexed['value'],
        title='Monthly anti-diabetic drug sales in Australia from 1992 to 2008.')

# Importing the time series data for Air Passengers
df_air_passengers = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/AirPassengers.csv', parse_dates=['date'])
x = df_air_passengers['date'].values
y1 = df_air_passengers['value'].values

# Plot Air Passengers (Two Side View)
fig, ax = plt.subplots(1, 1, figsize=(16, 5), dpi=120)
plt.fill_between(x, y1=y1, y2=-y1, alpha=0.5, linewidth=2, color='seagreen')
plt.ylim(-800, 800)
plt.title('Air Passengers (Two Side View)', fontsize=16)
plt.hlines(y=0, xmin=np.min(df_air_passengers.date), xmax=np.max(df_air_passengers.date), linewidth=.5)
plt.show()

# Seasonal Plot of Drug Sales
df_a10 = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/a10.csv', parse_dates=['date'], index_col='date')
df_a10.reset_index(inplace=True)

# Prepare data for seasonal plot
df_a10['year'] = [d.year for d in df_a10.date]
df_a10['month'] = [d.strftime('%b') for d in df_a10.date]
years = df_a10['year'].unique()

# Random colors for each year
np.random.seed(100)
mycolors = np.random.choice(list(mpl.colors.XKCD_COLORS.keys()), len(years), replace=False)

# Draw Seasonal Plot
plt.figure(figsize=(16, 12), dpi=80)
for i, y in enumerate(years):
    if i > 0:
        plt.plot('month', 'value', data=df_a10.loc[df_a10.year == y, :], color=mycolors[i], label=y)
        plt.text(df_a10.loc[df_a10.year == y, :].shape[0] - .9, df_a10.loc[df_a10.year == y, 'value'][-1:].values[0], y, fontsize=12, color=mycolors[i])

# Decorations
plt.gca().set(xlim=(-0.3, 11), ylim=(2, 30), ylabel='$Drug Sales$', xlabel='$Month$')
plt.yticks(fontsize=12, alpha=.7)
plt.title("Seasonal Plot of Drug Sales Time Series", fontsize=20)
plt.show()

# Year-wise and Month-wise Box Plot
fig, axes = plt.subplots(1, 2, figsize=(20, 7), dpi=80)
sns.boxplot(x='year', y='value', data=df_a10, ax=axes[0])
sns.boxplot(x='month', y='value', data=df_a10.loc[~df_a10.year.isin([1991, 2008]), :], ax=axes[1])

# Set Title
axes[0].set_title('Year-wise Box Plot\n(The Trend)', fontsize=18)
axes[1].set_title('Month-wise Box Plot\n(The Seasonality)', fontsize=18)
plt.show()
```
